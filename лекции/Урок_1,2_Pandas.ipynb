### Создание Series
**Series** - одна из структур данных библиотеки Pandas. Похожа на словарь, но является упорядоченной

```python
a = np.array([2, 5, 7, 8, 3]) # можно просто список
```

```python
b = pd.Series(a)
b # Левый столбец - индексы, правый - значений 
```

можно задать индексы явно(числа или символы)

```python
b = pd.Series(a, index = [0, 1, 3, 2, 4]) # обрати внимание, список не сортируется по индексу
b
```

```python
b = pd.Series(a, index = ['a', 'b', 'c', 'd', 'e'])
b
```
