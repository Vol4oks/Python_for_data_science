```python
import numpy as np
```

### Выбор данных

```python
a = np.random.randint(0, 20, 10)
a
```

```python
a[a > 10]
```

```python
a[(a > 10) & (a % 2 == 0)]
```

```python
a[(a > 10) | (a % 3 == 0)]
```

**numpy.where** - возвращает индексы нужных чисел 

```python
a 
```

```python
# Возвращение индекса чисел, удовлетворяющих условию
np.where(a % 2 == 0)
```

```python
a[np.where(a % 2 == 0)]
```

```python
a = [1, 4, 7]
b = [9, 2, 5]
```

```python
# True = берет значение из первого списка
# Faulse = из второго
np.where([False, True, True], a, b)
```

### Сортировка

```python
a = np.random.randint(0, 10, size=(3, 4))
a
```

```python
a[[2, 0, 1], :]
```

```python
a.argsort(axis = 0) # получаем индексы
```

```python
a[:, 0].argsort()
```

```python
a[a[:, 0].argsort(), :]
```
